## Função lambda em Python
---

###
Vamos usar uma função `lambda` para representar diretamente uma função matemática.

**Estrutura:**
```python
lambda x: lei da função
```

Interpretação matemática:

$f(x) = -3x + 7$

```python
f = lambda x: -3*x + 7
print(f(5))
```

A leitura é direta:

$x \mapsto 2x + 3$

### Exemplos de códigos com algumas funções matemáticas

**Função Afim**

$f(x) = -3x + 7$

````python
f = lambda x: -3*x + 7
print(f(5))
````

**Função quadrática**

$h(x) = x^2 - 4x + 1$

````python
h = lambda x: x**2 - 4*x + 1
print(h(2))
````

**Função cúbica**

$g(x) = x^3$

````python
g = lambda x: x**3
print(g(3))
````

## Matplotlib
---

O matplotlib permite visualizar graficamente a relação entre:

- domínio (valores de $x$)
- imagem (valores de $f(x)$)

Ou seja:

$
\begin{aligned}
f: A &\to B \\
x &\mapsto f(x)
\end{aligned}
$


### Exemplo passo a passo do uso da Matplotlib

**Passo 1**: Lei da função com a função lambda

````python

# Definição da função matemática

f = lambda x: x**2 - 4*x + 1
````

**Passo 2**: Domínio da função com np.linspace

````python

import numpy as np

# np.linspace(inicio, fim, quantidade)
# Gera vários valores igualmente espaçados entre inicio e fim

# Aqui nesse exemplo:
# -2 é o início do domínio
# 6 é o fim do domínio
# 100 é a quantidade de pontos

x = np.linspace(-2, 6, 100)

# Isso representa vários valores de x no intervalo escolhido
````

**Passo 3**: mostrar o gráfico usando a biblioteca matplotlib

````python
import matplotlib.pyplot as plt

# Calcula os valores de y aplicando a função em cada x
y = f(x)

# Cria o gráfico ligando os pontos (x, y)
plt.plot(x, y)

# Título do gráfico (representa a função estudada)
plt.title("f(x) = x^2 - 4x + 1")

# Nome do eixo horizontal (domínio)
plt.xlabel("x")

# Nome do eixo vertical (imagem da função)
plt.ylabel("f(x)")

# Ativa grade para facilitar leitura dos pontos
plt.grid()

# Exibe o gráfico
plt.show()
````
